```python
import pandas as pd
import numpy as np
import pickle
```

```python
df_t20 = pickle.load(open('dataset_t20_3.pkl','rb'))
```

```python
df =df_t20.copy()
```

## <font color='blue'>Extracting City </font>

```python
df_t20.isnull().sum()
```

```python
df_t20[df_t20['city'].isnull()]['venue'].value_counts()
```

```python
cities = np.where(df_t20['city'].isnull(),df_t20['venue'].str.split().apply(lambda x:x[0]),df_t20['city'])
```

```python
print(len(cities))
```

```python
df['city']=cities 
```

```python
df.isnull().sum()
```

<font color='red'> Now we resolved the null values of city and also get a city for all the columns .Hence their is no null values in the datasets</font>

```python
df.drop(columns=['venue'],inplace=True)
```

#### Now we are taking cities only in which more than 5 matches or600 balls is played .

```python
eligible_cities = df['city'].value_counts()[df['city'].value_counts() > 600].index.tolist()
```

```python
len(eligible_cities)
```

```python
eligible_cities
```

```python
df = df[df['city'].isin(eligible_cities)]
```

```python
df.shape
```

### <font color='blue'>Extracting Current Score</font> 

```python
df.groupby('match_id').cumsum()['runs']
```

```python
df['current_score'] = df.groupby('match_id').cumsum()['runs']
```

```python
df.head()
```

### <font color='blue'>Extracting Ball Left

```python
df['over'] = df['ball'].apply(lambda x:str(x).split('.')[0])
```

```python
df['ball_no'] = df['ball'].apply(lambda x : str(x).split('.')[1])
```

```python
df['balls_bowled'] = (df['over'].astype('int')*6) + (df['ball_no'].astype('int'))
```

```python
df['over'].astype('int')*6
```

```python
df['ball_left'] = 120-df['balls_bowled']
```

```python
#Some inning have more than 120 balls due to extras
df['ball_left'] = df['ball_left'].apply(lambda x : 0 if x<0 else x)
```

```python
df.head()
```

### <FONT color='blue'>Extracting Wicket left

```python
df['player_dismissed'] = df['player_of_dismissed'].apply(lambda x: 0 if x=='0' else 1)
df['player_dismissed'] = df['player_dismissed'].astype('int')
```

```python
df['player_dismissed'] = df.groupby('match_id').cumsum()['player_dismissed']
```

```python
df['wicket_left'] = 10 -df['player_dismissed']
```

```python
df.drop(columns=['player_of_dismissed','player_dismissed'],inplace=True)
```

```python
df
```

### <font color='blue'> Extracting Current RunRate

```python
df['curr_rr'] = (df['current_score']*6)/df['balls_bowled']
```

```python
df.head(10)
```

### <font color='blue'> Extracting Last Five Over Runs</font>

```python
groups = df.groupby('match_id')
match_ids = df['match_id'].unique()
```

```python
len(match_ids)
```

```python
groups.get_group(2).rolling(window=30).sum()['runs'].values[30:]
```

```python
last_five= []
for ind in match_ids:
    
    last_five.extend(groups.get_group(ind).rolling(window=30).sum()['runs'].values.tolist())
```

```python
df['last_five']=last_five
```

```python
df.iloc[30:]
```

### <font color='green'>Extracting Final Score

```python
df.groupby('match_id').sum()['runs']
```

```python
df = df.groupby('match_id').sum()['runs'].reset_index().merge(df,on='match_id')
```

```python
df.head(5)
```

<font color='red'> run_x is total score and run_y is run score on that ball

```python
df_main = df[['batting_team','bowling_team','current_score','wicket_left','curr_rr','city','ball_left','last_five','runs_x']]
```

```python
df_main.isnull().sum()
```

```python
df_main.dropna(inplace=True)
```

```python
df_main.isnull().sum()
```

```python
df_main = df_main.sample(df_main.shape[0])
```
